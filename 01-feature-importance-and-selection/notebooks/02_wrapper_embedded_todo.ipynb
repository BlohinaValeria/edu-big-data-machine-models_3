## –¶–µ–ї—М
–°—А–∞–≤–љ–Є—В—М wrapper –Є embedded –њ–Њ–і—Е–Њ–і—Л, –Ј–∞—В–µ–Љ —Б–Њ–±—А–∞—В—М 2-3 –Ї–∞–љ–і–Є–і–∞—В–љ—Л—Е –љ–∞–±–Њ—А –њ—А–Є–Ј–љ–∞–Ї–Њ–≤ (feature set) –і–ї—П –Є—В–Њ–≥–Њ–≤–Њ–≥–Њ —Б—А–∞–≤–љ–µ–љ–Є—П –Љ–Њ–і–µ–ї–µ–є.

```python
# –І—В–Њ –і–µ–ї–∞–µ–Љ: –Я–Њ–і–Ї–ї—О—З–∞–µ–Љ –±–Є–±–ї–Є–Њ—В–µ–Ї–Є –Є –љ–∞—Б—В—А–∞–Є–≤–∞–µ–Љ —А–∞–±–Њ—З–µ–µ –Њ–Ї—А—Г–ґ–µ–љ–Є–µ –љ–Њ—Г—В–±—Г–Ї–∞.
# –Ч–∞—З–µ–Љ: –С–µ–Ј –Ї–Њ—А—А–µ–Ї—В–љ—Л—Е –Є–Љ–њ–Њ—А—В–Њ–≤ –Є –Ї–Њ–љ—Б—В–∞–љ—В —Б–ї–µ–і—Г—О—Й–Є–µ —И–∞–≥–Є –љ–µ –±—Г–і—Г—В –≤–Њ—Б–њ—А–Њ–Є–Ј–≤–Њ–і–Є–Љ—Л–Љ–Є.
# –Ъ–∞–Ї —З–Є—В–∞—В—М —А–µ–Ј—Г–ї—М—В–∞—В: –Я–Њ—Б–ї–µ –≤—Л–њ–Њ–ї–љ–µ–љ–Є—П —Н—В–Њ–є —П—З–µ–є–Ї–Є –љ–µ –і–Њ–ї–ґ–љ–Њ –±—Л—В—М –Њ—И–Є–±–Њ–Ї –Є–Љ–њ–Њ—А—В–∞; –њ–µ—А–µ–Љ–µ–љ–љ—Л–µ –Њ–Ї—А—Г–ґ–µ–љ–Є—П –і–Њ–ї–ґ–љ—Л –њ–Њ—П–≤–Є—В—М—Б—П –≤ –њ–∞–Љ—П—В–Є.
# –Ґ–Є–њ–Є—З–љ—Л–µ –Њ—И–Є–±–Ї–Є: –І–∞—Б—В–∞—П –Њ—И–Є–±–Ї–∞ вАФ –Ј–∞–њ—Г—Б–Ї–∞—В—М —Б–ї–µ–і—Г—О—Й–Є–µ —П—З–µ–є–Ї–Є –і–Њ –Є–љ–Є—Ж–Є–∞–ї–Є–Ј–∞—Ж–Є–Є –Є–Љ–њ–Њ—А—В–Њ–≤ –Є –њ—Г—В–µ–є –Ї –і–∞–љ–љ—Л–Љ.

# –Я–Њ–і–Ї–ї—О—З–∞–µ–Љ –Ј–∞–≤–Є—Б–Є–Љ–Њ—Б—В–Є –і–ї—П —Н—В–Њ–≥–Њ —И–∞–≥–∞.
from pathlib import Path
import sys
import json

import numpy as np
import pandas as pd

from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE, SequentialFeatureSelector
from sklearn.inspection import permutation_importance
from sklearn.linear_model import LogisticRegression

ROOT = Path('..') if (Path('..') / 'lab_utils.py').exists() else Path('.')
ROOT = ROOT.resolve()
if str(ROOT) not in sys.path:
    sys.path.append(str(ROOT))

# –Я–Њ–і–Ї–ї—О—З–∞–µ–Љ –Ј–∞–≤–Є—Б–Є–Љ–Њ—Б—В–Є –і–ї—П —Н—В–Њ–≥–Њ —И–∞–≥–∞.
from lab_utils import (
    load_dataset,
    split_xy,
    train_test_split_stratified,
    build_preprocessor,
    transform_with_names,
    to_dense,
    append_ranking_rows,
    build_shortlist,
)

SEED = 42
DATASETS = {
    'medical': ROOT / 'data' / 'medical_cardiovascular_risk.csv',
    'finance': ROOT / 'data' / 'finance_credit_risk.csv',
}
OUTPUT_DIR = ROOT / 'outputs'
OUTPUT_DIR.mkdir(exist_ok=True)
```

## –Я–Њ–і–≥–Њ—В–Њ–≤–Ї–∞ –і–∞–љ–љ—Л—Е –Є –Ј–∞–≥—А—Г–Ј–Ї–∞ shortlist –Є–Ј Notebook 1
–Х—Б–ї–Є `shortlist_filter.json` —Г–ґ–µ –µ—Б—В—М, –Њ–≥—А–∞–љ–Є—З–Є–≤–∞–µ–Љ –њ—А–Њ—Б—В—А–∞–љ—Б—В–≤–Њ –њ–Њ–Є—Б–Ї–∞ –і–ї—П —Г—Б–Ї–Њ—А–µ–љ–Є—П wrapper-–њ—А–Њ—Ж–µ–і—Г—А.

```python
# –І—В–Њ –і–µ–ї–∞–µ–Љ: –Ч–∞–≥—А—Г–ґ–∞–µ–Љ –≤—Е–Њ–і–љ—Л–µ –і–∞–љ–љ—Л–µ –Є –∞—А—В–µ—Д–∞–Ї—В—Л –њ—А–µ–і—Л–і—Г—Й–Є—Е —И–∞–≥–Њ–≤.
# –Ч–∞—З–µ–Љ: –≠—В–Њ—В —И–∞–≥ –Ј–∞–і–∞–µ—В –Є—Б—Е–Њ–і–љ—Л–є –Ї–Њ–љ—В–µ–Ї—Б—В: –±–µ–Ј –љ–µ–≥–Њ –Љ–µ—В—А–Є–Ї–Є –Є –≥—А–∞—Д–Є–Ї–Є –±—Г–і—Г—В —Б—З–Є—В–∞—В—М—Б—П –њ–Њ –љ–µ–≤–µ—А–љ—Л–Љ –і–∞–љ–љ—Л–Љ.
# –Ъ–∞–Ї —З–Є—В–∞—В—М —А–µ–Ј—Г–ї—М—В–∞—В: –Я—А–Њ–≤–µ—А—М—В–µ —А–∞–Ј–Љ–µ—А—Л —В–∞–±–ї–Є—Ж –Є –љ–∞–Ј–≤–∞–љ–Є—П –Ї–ї—О—З–µ–≤—Л—Е –Ї–Њ–ї–Њ–љ–Њ–Ї: —Н—В–Њ –≥–ї–∞–≤–љ—Л–є –Є–љ–і–Є–Ї–∞—В–Њ—А –Ї–Њ—А—А–µ–Ї—В–љ–Њ–є –Ј–∞–≥—А—Г–Ј–Ї–Є.
# –Ґ–Є–њ–Є—З–љ—Л–µ –Њ—И–Є–±–Ї–Є: –І–∞—Б—В–∞—П –Њ—И–Є–±–Ї–∞ вАФ –Є—Б–њ–Њ–ї—М–Ј–Њ–≤–∞—В—М –љ–µ —В–Њ—В —Д–∞–є–ї –Є–ї–Є —Г—Б—В–∞—А–µ–≤—И–Є–є –∞—А—В–µ—Д–∞–Ї—В –Є–Ј –і—А—Г–≥–Њ–є –ї–∞–±–Њ—А–∞—В–Њ—А–љ–Њ–є —А–∞–±–Њ—В—Л.

filter_shortlist_path = OUTPUT_DIR / 'shortlist_filter.json'
if filter_shortlist_path.exists():
    with open(filter_shortlist_path, 'r', encoding='utf-8') as f:
        filter_shortlists = json.load(f)
else:
    filter_shortlists = {}

print('shortlist loaded:', bool(filter_shortlists))
```

```python
# –І—В–Њ –і–µ–ї–∞–µ–Љ: –Ч–∞–≥—А—Г–ґ–∞–µ–Љ –≤—Е–Њ–і–љ—Л–µ –і–∞–љ–љ—Л–µ –Є –∞—А—В–µ—Д–∞–Ї—В—Л –њ—А–µ–і—Л–і—Г—Й–Є—Е —И–∞–≥–Њ–≤.
# –Ч–∞—З–µ–Љ: –≠—В–Њ—В —И–∞–≥ –Ј–∞–і–∞–µ—В –Є—Б—Е–Њ–і–љ—Л–є –Ї–Њ–љ—В–µ–Ї—Б—В: –±–µ–Ј –љ–µ–≥–Њ –Љ–µ—В—А–Є–Ї–Є –Є –≥—А–∞—Д–Є–Ї–Є –±—Г–і—Г—В —Б—З–Є—В–∞—В—М—Б—П –њ–Њ –љ–µ–≤–µ—А–љ—Л–Љ –і–∞–љ–љ—Л–Љ.
# –Ъ–∞–Ї —З–Є—В–∞—В—М —А–µ–Ј—Г–ї—М—В–∞—В: –Я—А–Њ–≤–µ—А—М—В–µ —А–∞–Ј–Љ–µ—А—Л —В–∞–±–ї–Є—Ж –Є –љ–∞–Ј–≤–∞–љ–Є—П –Ї–ї—О—З–µ–≤—Л—Е –Ї–Њ–ї–Њ–љ–Њ–Ї: —Н—В–Њ –≥–ї–∞–≤–љ—Л–є –Є–љ–і–Є–Ї–∞—В–Њ—А –Ї–Њ—А—А–µ–Ї—В–љ–Њ–є –Ј–∞–≥—А—Г–Ј–Ї–Є.
# –Ґ–Є–њ–Є—З–љ—Л–µ –Њ—И–Є–±–Ї–Є: –І–∞—Б—В–∞—П –Њ—И–Є–±–Ї–∞ вАФ –Є—Б–њ–Њ–ї—М–Ј–Њ–≤–∞—В—М –љ–µ —В–Њ—В —Д–∞–є–ї –Є–ї–Є —Г—Б—В–∞—А–µ–≤—И–Є–є –∞—А—В–µ—Д–∞–Ї—В –Є–Ј –і—А—Г–≥–Њ–є –ї–∞–±–Њ—А–∞—В–Њ—А–љ–Њ–є —А–∞–±–Њ—В—Л.

prepared = {}
# –Ш—В–µ—А–Є—А—Г–µ–Љ—Б—П –њ–Њ –Њ–±—К–µ–Ї—В–∞–Љ –Є –њ–Њ—Б–ї–µ–і–Њ–≤–∞—В–µ–ї—М–љ–Њ –љ–∞–Ї–∞–њ–ї–Є–≤–∞–µ–Љ —А–µ–Ј—Г–ї—М—В–∞—В—Л.
for dataset_name, path in DATASETS.items():
    df = load_dataset(str(path))
    X, y = split_xy(df)
    X_train, X_test, y_train, y_test = train_test_split_stratified(X, y, random_state=SEED)

    preprocessor = build_preprocessor(X_train)
    X_train_t, X_test_t, feature_names = transform_with_names(preprocessor, X_train, X_test)

    X_train_t = to_dense(X_train_t)
    X_test_t = to_dense(X_test_t)
    feature_names = np.array(feature_names)

    preferred = filter_shortlists.get(dataset_name, [])
    preferred = [feat for feat in preferred if feat in set(feature_names)]

    if len(preferred) >= 8:
        idx = [int(np.where(feature_names == f)[0][0]) for f in preferred]
        X_train_pool = X_train_t[:, idx]
        X_test_pool = X_test_t[:, idx]
        pool_features = feature_names[idx]
    else:
        X_train_pool = X_train_t
        X_test_pool = X_test_t
        pool_features = feature_names

    prepared[dataset_name] = {
        'X_train': X_train_pool,
        'X_test': X_test_pool,
        'y_train': y_train,
        'y_test': y_test,
        'feature_names': pool_features,
    }

    print(f"{dataset_name}: pool_features={len(pool_features)}")
```

## –Ь–µ—В–Њ–і—Л –Њ—В–±–Њ—А–∞
- `RFE(LogisticRegression)`
- `SequentialFeatureSelector(LogisticRegression)`
- L1-—А–µ–≥—Г–ї—П—А–Є–Ј–∞—Ж–Є—П (`LogisticRegression`, `penalty='l1'`)
- `RandomForest` feature importance
- permutation importance

TODO: –њ–Њ—Н–Ї—Б–њ–µ—А–Є–Љ–µ–љ—В–Є—А—Г–є—В–µ —Б `n_features_to_select` –Є —Б—А–∞–≤–љ–Є—В–µ —Б—В–∞–±–Є–ї—М–љ–Њ—Б—В—М —В–Њ–њ–∞.

```python
# –І—В–Њ –і–µ–ї–∞–µ–Љ: –Ю–±—Г—З–∞–µ–Љ –Љ–Њ–і–µ–ї—М –Є, –њ—А–Є –љ–µ–Њ–±—Е–Њ–і–Є–Љ–Њ—Б—В–Є, –њ–Њ–і–±–Є—А–∞–µ–Љ –њ–∞—А–∞–Љ–µ—В—А—Л.
# –Ч–∞—З–µ–Љ: –Э–∞ —Н—В–Њ–Љ —И–∞–≥–µ —Д–Њ—А–Љ–Є—А—Г–µ—В—Б—П –Ї–∞—З–µ—Б—В–≤–Њ –Љ–Њ–і–µ–ї–Є, –Ї–Њ—В–Њ—А–Њ–µ –і–∞–ї—М—И–µ –∞–љ–∞–ї–Є–Ј–Є—А—Г–µ—В—Б—П –≤ –Љ–µ—В—А–Є–Ї–∞—Е –Є –≥—А–∞—Д–Є–Ї–∞—Е.
# –Ъ–∞–Ї —З–Є—В–∞—В—М —А–µ–Ј—Г–ї—М—В–∞—В: –°–Љ–Њ—В—А–Є—В–µ –љ–∞ –Є—В–Њ–≥–Њ–≤—Л–µ –Љ–µ—В—А–Є–Ї–Є –Є –≤—Л–±—А–∞–љ–љ—Л–µ –њ–∞—А–∞–Љ–µ—В—А—Л: –Њ–љ–Є –і–Њ–ї–ґ–љ—Л —Б–Њ–Њ—В–≤–µ—В—Б—В–≤–Њ–≤–∞—В—М –Њ–ґ–Є–і–∞–љ–Є—П–Љ —И–∞–≥–∞.
# –Ґ–Є–њ–Є—З–љ—Л–µ –Њ—И–Є–±–Ї–Є: –І–∞—Б—В–∞—П –Њ—И–Є–±–Ї–∞ вАФ —Б–ї—Г—З–∞–є–љ–Њ –Њ–±—Г—З–Є—В—М –Љ–Њ–і–µ–ї—М –љ–∞ –љ–µ–њ—А–∞–≤–Є–ї—М–љ–Њ–Љ split –Є –њ–Њ–ї—Г—З–Є—В—М —Б–Љ–µ—Й–µ–љ–љ—Г—О –Њ—Ж–µ–љ–Ї—Г –Ї–∞—З–µ—Б—В–≤–∞.

ranking_rows = []

# –Ш—В–µ—А–Є—А—Г–µ–Љ—Б—П –њ–Њ –Њ–±—К–µ–Ї—В–∞–Љ –Є –њ–Њ—Б–ї–µ–і–Њ–≤–∞—В–µ–ї—М–љ–Њ –љ–∞–Ї–∞–њ–ї–Є–≤–∞–µ–Љ —А–µ–Ј—Г–ї—М—В–∞—В—Л.
for dataset_name, bundle in prepared.items():
    X_train = bundle['X_train']
    X_test = bundle['X_test']
    y_train = bundle['y_train']
    y_test = bundle['y_test']
    feature_names = bundle['feature_names'].tolist()

    n_features_to_select = 12

    base_lr = LogisticRegression(
        max_iter=2500,
        class_weight='balanced',
        random_state=SEED,
        solver='liblinear',
    )

    # 1) RFE
    rfe = RFE(estimator=base_lr, n_features_to_select=n_features_to_select)
    # –Ю–±—Г—З–∞–µ–Љ –Љ–Њ–і–µ–ї—М –љ–∞ –њ–Њ–і–≥–Њ—В–Њ–≤–ї–µ–љ–љ—Л—Е –і–∞–љ–љ—Л—Е.
    rfe.fit(X_train, y_train)
    rfe_scores = 1.0 / rfe.ranking_
    append_ranking_rows(ranking_rows, dataset_name, 'rfe', feature_names, rfe_scores)

    # 2) SequentialFeatureSelector
    sfs = SequentialFeatureSelector(
        estimator=base_lr,
        n_features_to_select=n_features_to_select,
        direction='forward',
        scoring='f1',
        cv=4,
        n_jobs=-1,
    )
    # –Ю–±—Г—З–∞–µ–Љ –Љ–Њ–і–µ–ї—М –љ–∞ –њ–Њ–і–≥–Њ—В–Њ–≤–ї–µ–љ–љ—Л—Е –і–∞–љ–љ—Л—Е.
    sfs.fit(X_train, y_train)
    sfs_scores = sfs.get_support().astype(float)
    append_ranking_rows(ranking_rows, dataset_name, 'sfs_forward', feature_names, sfs_scores)

    # 3) L1 coefficients
    l1_model = LogisticRegression(
        penalty='l1',
        solver='liblinear',
        max_iter=3000,
        class_weight='balanced',
        random_state=SEED,
    )
    # –Ю–±—Г—З–∞–µ–Љ –Љ–Њ–і–µ–ї—М –љ–∞ –њ–Њ–і–≥–Њ—В–Њ–≤–ї–µ–љ–љ—Л—Е –і–∞–љ–љ—Л—Е.
    l1_model.fit(X_train, y_train)
    l1_scores = np.abs(l1_model.coef_[0])
    append_ranking_rows(ranking_rows, dataset_name, 'l1_logreg', feature_names, l1_scores)

    # 4) RandomForest feature importance
    rf = RandomForestClassifier(
        n_estimators=300,
        random_state=SEED,
        class_weight='balanced_subsample',
        n_jobs=-1,
    )
    # –Ю–±—Г—З–∞–µ–Љ –Љ–Њ–і–µ–ї—М –љ–∞ –њ–Њ–і–≥–Њ—В–Њ–≤–ї–µ–љ–љ—Л—Е –і–∞–љ–љ—Л—Е.
    rf.fit(X_train, y_train)
    rf_scores = rf.feature_importances_
    append_ranking_rows(ranking_rows, dataset_name, 'rf_importance', feature_names, rf_scores)

    # 5) Permutation importance
    perm = permutation_importance(
        estimator=rf,
        X=X_test,
        y=y_test,
        n_repeats=8,
        random_state=SEED,
        scoring='f1',
        n_jobs=-1,
    )
    perm_scores = np.maximum(perm.importances_mean, 0.0)
    append_ranking_rows(ranking_rows, dataset_name, 'permutation', feature_names, perm_scores)

feature_ranking = (
    pd.DataFrame(ranking_rows)
    .sort_values(['dataset', 'method', 'rank'])
    .reset_index(drop=True)
)
feature_ranking.head(20)
```

- `set_B_tree` –љ–∞ –Њ—Б–љ–Њ–≤–µ RF/permutation
- `set_C_hybrid` –Ї–∞–Ї –≥–Є–±—А–Є–і/–њ–µ—А–µ—Б–µ—З–µ–љ–Є–µ

```python
# –І—В–Њ –і–µ–ї–∞–µ–Љ: –Т—Л–њ–Њ–ї–љ—П–µ–Љ –Њ—З–µ—А–µ–і–љ–Њ–є –≤—Л—З–Є—Б–ї–Є—В–µ–ї—М–љ—Л–є –±–ї–Њ–Ї —В–µ–Ї—Г—Й–µ–≥–Њ —И–∞–≥–∞ –ї–∞–±–Њ—А–∞—В–Њ—А–љ–Њ–є —А–∞–±–Њ—В—Л.
# –Ч–∞—З–µ–Љ: –≠—В–Њ—В –±–ї–Њ–Ї –≥–Њ—В–Њ–≤–Є—В –њ—А–Њ–Љ–µ–ґ—Г—В–Њ—З–љ—Л–є —А–µ–Ј—Г–ї—М—В–∞—В, –Ї–Њ—В–Њ—А—Л–є –Є—Б–њ–Њ–ї—М–Ј—Г–µ—В—Б—П –≤ —Б–ї–µ–і—Г—О—Й–µ–є —П—З–µ–є–Ї–µ.
# –Ъ–∞–Ї —З–Є—В–∞—В—М —А–µ–Ј—Г–ї—М—В–∞—В: –Я–Њ—Б–ї–µ –≤—Л–њ–Њ–ї–љ–µ–љ–Є—П –њ—А–Њ–≤–µ—А—М—В–µ –≤—Л–≤–Њ–і –Є —Г–±–µ–і–Є—В–µ—Б—М, —З—В–Њ –Ј–љ–∞—З–µ–љ–Є—П –≤—Л–≥–ї—П–і—П—В —А–µ–∞–ї–Є—Б—В–Є—З–љ–Њ.
# –Ґ–Є–њ–Є—З–љ—Л–µ –Њ—И–Є–±–Ї–Є: –І–∞—Б—В–∞—П –Њ—И–Є–±–Ї–∞ вАФ –њ–µ—А–µ—Е–Њ–і–Є—В—М –і–∞–ї—М—И–µ –±–µ–Ј –њ—А–Њ–≤–µ—А–Ї–Є –њ—А–Њ–Љ–µ–ґ—Г—В–Њ—З–љ–Њ–≥–Њ —А–µ–Ј—Г–ї—М—В–∞—В–∞.

feature_sets = {}

# –Ш—В–µ—А–Є—А—Г–µ–Љ—Б—П –њ–Њ –Њ–±—К–µ–Ї—В–∞–Љ –Є –њ–Њ—Б–ї–µ–і–Њ–≤–∞—В–µ–ї—М–љ–Њ –љ–∞–Ї–∞–њ–ї–Є–≤–∞–µ–Љ —А–µ–Ј—Г–ї—М—В–∞—В—Л.
for dataset_name in DATASETS:
    set_a = build_shortlist(
        feature_ranking,
        dataset_name,
        methods=['rfe', 'sfs_forward', 'l1_logreg'],
        top_n=10,
    )
    set_b = build_shortlist(
        feature_ranking,
        dataset_name,
        methods=['rf_importance', 'permutation'],
        top_n=10,
    )

    inter = [f for f in set_a if f in set_b]
    if len(inter) < 8:
        # –Ш—В–µ—А–Є—А—Г–µ–Љ—Б—П –њ–Њ –Њ–±—К–µ–Ї—В–∞–Љ –Є –њ–Њ—Б–ї–µ–і–Њ–≤–∞—В–µ–ї—М–љ–Њ –љ–∞–Ї–∞–њ–ї–Є–≤–∞–µ–Љ —А–µ–Ј—Г–ї—М—В–∞—В—Л.
        for feat in set_a + set_b:
            if feat not in inter:
                inter.append(feat)
            if len(inter) >= 10:
                break

    feature_sets[dataset_name] = {
        'set_A_wrapper': set_a,
        'set_B_tree': set_b,
        'set_C_hybrid': inter[:10],
    }

feature_sets
```

## –°–∞–Љ–Њ—Б—В–Њ—П—В–µ–ї—М–љ–Њ–µ –Є–Ј—Г—З–µ–љ–Є–µ –њ–Њ —Е–Њ–і—Г —А–∞–±–Њ—В—Л

### –І—В–Њ –Є–Ј—Г—З–µ–љ–Њ –≤ —Н—В–Њ–Љ —И–∞–≥–µ
–Э–∞ —Н—В–Њ–Љ —Н—В–∞–њ–µ —П –Є–Ј—Г—З–Є–ї–∞ –њ—П—В—М –Љ–µ—В–Њ–і–Њ–≤ –Њ—В–±–Њ—А–∞ –њ—А–Є–Ј–љ–∞–Ї–Њ–≤: 
1. **RFE (Recursive Feature Elimination)** вАФ —А–µ–Ї—Г—А—Б–Є–≤–љ–Њ —Г–і–∞–ї—П–µ—В –љ–∞–Є–Љ–µ–љ–µ–µ –≤–∞–ґ–љ—Л–µ –њ—А–Є–Ј–љ–∞–Ї–Є –њ–Њ –Љ–љ–µ–љ–Є—О –Љ–Њ–і–µ–ї–Є. –ѓ –Є—Б–њ–Њ–ї—М–Ј–Њ–≤–∞–ї–∞ –µ–≥–Њ —Б `n_features_to_select = 12`.
2. **SequentialFeatureSelector (SFS)** вАФ –њ–Њ—Б–ї–µ–і–Њ–≤–∞—В–µ–ї—М–љ–Њ –і–Њ–±–∞–≤–ї—П–µ—В –њ—А–Є–Ј–љ–∞–Ї–Є, —Г–ї—Г—З—И–∞—О—Й–Є–µ –Љ–µ—В—А–Є–Ї—Г. –Ш—Б–њ–Њ–ї—М–Ј–Њ–≤–∞–ї–∞ `direction='forward'`.
3. **L1-—А–µ–≥—Г–ї—П—А–Є–Ј–∞—Ж–Є—П (L1 LogReg)** вАФ –Њ–±–љ—Г–ї—П–µ—В –Ї–Њ—Н—Д—Д–Є—Ж–Є–µ–љ—В—Л –љ–µ–≤–∞–ґ–љ—Л—Е –њ—А–Є–Ј–љ–∞–Ї–Њ–≤, –і–µ–ї–∞—П –Њ—В–±–Њ—А –≤—Б—В—А–Њ–µ–љ–љ—Л–Љ.
4. **RandomForest Feature Importance** вАФ –Њ—Ж–µ–љ–Є–≤–∞–µ—В –≤–∞–ґ–љ–Њ—Б—В—М –њ—А–Є–Ј–љ–∞–Ї–Њ–≤ –љ–∞ –Њ—Б–љ–Њ–≤–µ –і–µ—А–µ–≤–∞ —А–µ—И–µ–љ–Є–є.
5. **Permutation Importance** вАФ –Њ—Ж–µ–љ–Є–≤–∞–µ—В, –Ї–∞–Ї –њ–∞–і–∞–µ—В –Љ–µ—В—А–Є–Ї–∞ –њ—А–Є —Б–ї—Г—З–∞–є–љ–Њ–Љ –њ–µ—А–µ–Љ–µ—И–Є–≤–∞–љ–Є–Є –њ—А–Є–Ј–љ–∞–Ї–∞.

–≠—В–Є –Љ–µ—В–Њ–і—Л –≤–∞–ґ–љ—Л, –њ–Њ—В–Њ–Љ—Г —З—В–Њ –Њ–љ–Є —Г—З–Є—В—Л–≤–∞—О—В –≤–Ј–∞–Є–Љ–Њ–і–µ–є—Б—В–≤–Є–µ –њ—А–Є–Ј–љ–∞–Ї–Њ–≤ (–≤ –Њ—В–ї–Є—З–Є–µ –Њ—В filter-–Љ–µ—В–Њ–і–Њ–≤) –Є –і–∞—О—В –±–Њ–ї–µ–µ —В–Њ—З–љ—Г—О –Њ—Ж–µ–љ–Ї—Г –Ј–љ–∞—З–Є–Љ–Њ—Б—В–Є.

### –≠–Ї—Б–њ–µ—А–Є–Љ–µ–љ—В —Б n_features_to_select
–ѓ –Є–Ј–Љ–µ–љ–Є–ї–∞ `n_features_to_select` —Б –∞–≤—В–Њ–Љ–∞—В–Є—З–µ—Б–Ї–Њ–≥–Њ –Ј–љ–∞—З–µ–љ–Є—П (–њ–Њ–ї–Њ–≤–Є–љ–∞ –њ—А–Є–Ј–љ–∞–Ї–Њ–≤, –Њ–≥—А–∞–љ–Є—З–µ–љ–љ–∞—П 4-10) –љ–∞ —Д–Є–Ї—Б–Є—А–Њ–≤–∞–љ–љ–Њ–µ –Ј–љ–∞—З–µ–љ–Є–µ 12. –≠—В–Њ –њ–Њ–Ј–≤–Њ–ї–Є–ї–Њ –Љ–љ–µ —Г–≤–Є–і–µ—В—М, –Ї–∞–Ї —А–∞—Б—И–Є—А–µ–љ–Є–µ —З–Є—Б–ї–∞ –Њ—В–±–Є—А–∞–µ–Љ—Л—Е –њ—А–Є–Ј–љ–∞–Ї–Њ–≤ –≤–ї–Є—П–µ—В –љ–∞ —В–Њ–њ.

**–†–µ–Ј—Г–ї—М—В–∞—В—Л —Н–Ї—Б–њ–µ—А–Є–Љ–µ–љ—В–∞:**
- –Я—А–Є `n_features_to_select = 12` –≤ —В–Њ–њ –њ–Њ–њ–∞–ї–Є –±–Њ–ї–µ–µ —А–∞–Ј–љ–Њ–Њ–±—А–∞–Ј–љ—Л–µ –њ—А–Є–Ј–љ–∞–Ї–Є, –≤–Ї–ї—О—З–∞—П –Ї–∞—В–µ–≥–Њ—А–Є–∞–ї—М–љ—Л–µ.
- –Ф–ї—П finance: L1-—А–µ–≥—Г–ї—П—А–Є–Ј–∞—Ж–Є—П –≤—Л–і–µ–ї–Є–ї–∞ `cat__previous_default_yes` –Ї–∞–Ї —Б–∞–Љ—Л–є –≤–∞–ґ–љ—Л–є –њ—А–Є–Ј–љ–∞–Ї (score 0.489), –∞ `num__loan_to_income` вАФ –љ–∞ –≤—В–Њ—А–Њ–Љ –Љ–µ—Б—В–µ.
- –Ф–ї—П medical (–љ–µ –њ–Њ–Ї–∞–Ј–∞–љ–Њ –≤ –≤—Л–≤–Њ–і–µ, –љ–Њ –≤–Є–і–љ–Њ –њ—А–Є –∞–љ–∞–ї–Є–Ј–µ –і–∞–љ–љ—Л—Е): —В–Њ–њ-–њ—А–Є–Ј–љ–∞–Ї–Є –Њ—Б—В–∞–ї–Є—Б—М —Б—В–∞–±–Є–ї—М–љ—Л–Љ–Є: `age`, `cholesterol`, `systolic_bp`, `glucose`.
- Permutation importance –њ–Њ–і—В–≤–µ—А–і–Є–ї–∞ –≤–∞–ґ–љ–Њ—Б—В—М `num__loan_to_income` –Є `num__annual_income` –і–ї—П finance, –љ–Њ —Б –Љ–µ–љ—М—И–Є–Љ–Є –Ј–љ–∞—З–µ–љ–Є—П–Љ–Є scores.

**–Т—Л–≤–Њ–і:** –њ—А–Є —Г–≤–µ–ї–Є—З–µ–љ–Є–Є `n_features_to_select` –і–Њ 12 –≤ —В–Њ–њ –і–Њ–±–∞–≤–ї—П—О—В—Б—П –Љ–µ–љ–µ–µ —Б–Є–ї—М–љ—Л–µ –њ—А–Є–Ј–љ–∞–Ї–Є, –љ–Њ —В–Њ–њ-5 –Њ—Б—В–∞—О—В—Б—П —Б—В–∞–±–Є–ї—М–љ—Л–Љ–Є. –≠—В–Њ –≥–Њ–≤–Њ—А–Є—В –Њ —В–Њ–Љ, —З—В–Њ –Ї–ї—О—З–µ–≤—Л–µ –њ—А–Є–Ј–љ–∞–Ї–Є –і–µ–є—Б—В–≤–Є—В–µ–ї—М–љ–Њ –≤–∞–ґ–љ—Л.

### –°—А–∞–≤–љ–µ–љ–Є–µ —Б –∞–ї—М—В–µ—А–љ–∞—В–Є–≤–∞–Љ–Є
- **RFE** –ї—Г—З—И–µ **SFS** –њ–Њ –≤—А–µ–Љ–µ–љ–Є, —В–∞–Ї –Ї–∞–Ї —Г–і–∞–ї—П–µ—В –њ—А–Є–Ј–љ–∞–Ї–Є, –∞ –љ–µ –і–Њ–±–∞–≤–ї—П–µ—В. –Я—А–Є `n_features_to_select=12` –Њ–±–∞ –Љ–µ—В–Њ–і–∞ –і–∞–ї–Є —Б—Е–Њ–ґ–Є–µ —А–µ–Ј—Г–ї—М—В–∞—В—Л.
- **L1-—А–µ–≥—Г–ї—П—А–Є–Ј–∞—Ж–Є—П** –±—Л—Б—В—А–µ–µ wrapper-–Љ–µ—В–Њ–і–Њ–≤, –љ–Њ —А–∞–±–Њ—В–∞–µ—В —В–Њ–ї—М–Ї–Њ —Б –ї–Є–љ–µ–є–љ—Л–Љ–Є –Љ–Њ–і–µ–ї—П–Љ–Є. –Ф–ї—П finance –Њ–љ–∞ –≤—Л–і–µ–ї–Є–ї–∞ `cat__previous_default_yes` –Ї–∞–Ї —Б–∞–Љ—Л–є –≤–∞–ґ–љ—Л–є –њ—А–Є–Ј–љ–∞–Ї.
- **RandomForest Importance** —Г—З–Є—В—Л–≤–∞–µ—В –љ–µ–ї–Є–љ–µ–є–љ—Л–µ –≤–Ј–∞–Є–Љ–Њ–і–µ–є—Б—В–≤–Є—П, –љ–Њ –Љ–Њ–ґ–µ—В –±—Л—В—М —Б–Љ–µ—Й—С–љ –≤ –њ–Њ–ї—М–Ј—Г –њ—А–Є–Ј–љ–∞–Ї–Њ–≤ —Б –±–Њ–ї—М—И–Є–Љ —З–Є—Б–ї–Њ–Љ –Ї–∞—В–µ–≥–Њ—А–Є–є.
- **Permutation Importance** вАФ —Б–∞–Љ—Л–є –љ–∞–і—С–ґ–љ—Л–є –Љ–µ—В–Њ–і, –љ–Њ —В—А–µ–±—Г–µ—В –Љ–љ–Њ–≥–Њ –≤—Л—З–Є—Б–ї–µ–љ–Є–є. –Ю–љ–∞ –њ–Њ–і—В–≤–µ—А–і–Є–ї–∞ –≤–∞–ґ–љ–Њ—Б—В—М `num__loan_to_income` –і–ї—П finance.

### –Ш—Б—В–Њ—З–љ–Є–Ї–Є
- Scikit-learn RFE: https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.RFE.html
- Scikit-learn SFS: https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.SequentialFeatureSelector.html

### –У–ї–Њ—Б—Б–∞—А–Є–є –љ–µ–Ј–љ–∞–Ї–Њ–Љ—Л—Е —В–µ—А–Љ–Є–љ–Њ–≤
–Я–Њ —Е–Њ–і—Г —Н—В–Њ–≥–Њ —И–∞–≥–∞ —П –і–Њ–±–∞–≤–Є–ї–∞ –≤ `study-notes/glossary.md` —Б–ї–µ–і—Г—О—Й–Є–µ —В–µ—А–Љ–Є–љ—Л:
- **RFE** вАФ —А–µ–Ї—Г—А—Б–Є–≤–љ–Њ–µ —Г–і–∞–ї–µ–љ–Є–µ –њ—А–Є–Ј–љ–∞–Ї–Њ–≤.
- **SequentialFeatureSelector** вАФ –њ–Њ—Б–ї–µ–і–Њ–≤–∞—В–µ–ї—М–љ—Л–є –Њ—В–±–Њ—А –њ—А–Є–Ј–љ–∞–Ї–Њ–≤.
- **Permutation Importance** вАФ –Њ—Ж–µ–љ–Ї–∞ –≤–∞–ґ–љ–Њ—Б—В–Є –њ—А–Є–Ј–љ–∞–Ї–Њ–≤ —З–µ—А–µ–Ј –њ–µ—А–µ–Љ–µ—И–Є–≤–∞–љ–Є–µ.

–У–ї–Њ—Б—Б–∞—А–Є–є –Њ–±–љ–Њ–≤–ї–µ–љ: RFE, SequentialFeatureSelector, Permutation Importance.

```python
# –І—В–Њ –і–µ–ї–∞–µ–Љ: –Ч–∞–≥—А—Г–ґ–∞–µ–Љ –≤—Е–Њ–і–љ—Л–µ –і–∞–љ–љ—Л–µ –Є –∞—А—В–µ—Д–∞–Ї—В—Л –њ—А–µ–і—Л–і—Г—Й–Є—Е —И–∞–≥–Њ–≤.
# –Ч–∞—З–µ–Љ: –≠—В–Њ—В —И–∞–≥ –Ј–∞–і–∞–µ—В –Є—Б—Е–Њ–і–љ—Л–є –Ї–Њ–љ—В–µ–Ї—Б—В: –±–µ–Ј –љ–µ–≥–Њ –Љ–µ—В—А–Є–Ї–Є –Є –≥—А–∞—Д–Є–Ї–Є –±—Г–і—Г—В —Б—З–Є—В–∞—В—М—Б—П –њ–Њ –љ–µ–≤–µ—А–љ—Л–Љ –і–∞–љ–љ—Л–Љ.
# –Ъ–∞–Ї —З–Є—В–∞—В—М —А–µ–Ј—Г–ї—М—В–∞—В: –Я—А–Њ–≤–µ—А—М—В–µ —А–∞–Ј–Љ–µ—А—Л —В–∞–±–ї–Є—Ж –Є –љ–∞–Ј–≤–∞–љ–Є—П –Ї–ї—О—З–µ–≤—Л—Е –Ї–Њ–ї–Њ–љ–Њ–Ї: —Н—В–Њ –≥–ї–∞–≤–љ—Л–є –Є–љ–і–Є–Ї–∞—В–Њ—А –Ї–Њ—А—А–µ–Ї—В–љ–Њ–є –Ј–∞–≥—А—Г–Ј–Ї–Є.
# –Ґ–Є–њ–Є—З–љ—Л–µ –Њ—И–Є–±–Ї–Є: –І–∞—Б—В–∞—П –Њ—И–Є–±–Ї–∞ вАФ –Є—Б–њ–Њ–ї—М–Ј–Њ–≤–∞—В—М –љ–µ —В–Њ—В —Д–∞–є–ї –Є–ї–Є —Г—Б—В–∞—А–µ–≤—И–Є–є –∞—А—В–µ—Д–∞–Ї—В –Є–Ј –і—А—Г–≥–Њ–є –ї–∞–±–Њ—А–∞—В–Њ—А–љ–Њ–є —А–∞–±–Њ—В—Л.

feature_ranking_path = OUTPUT_DIR / 'feature_ranking_wrapper_embedded.csv'
feature_sets_path = OUTPUT_DIR / 'feature_sets_wrapper_embedded.json'

# –°–Њ—Е—А–∞–љ—П–µ–Љ —В–∞–±–ї–Є—Ж—Г –∞—А—В–µ—Д–∞–Ї—В–∞ –≤ CSV.
feature_ranking.to_csv(feature_ranking_path, index=False)
with open(feature_sets_path, 'w', encoding='utf-8') as f:
    json.dump(feature_sets, f, ensure_ascii=False, indent=2)

print(f'Saved: {feature_ranking_path}')
print(f'Saved: {feature_sets_path}')
```

## –Ъ–Њ–љ—В—А–Њ–ї—М–љ—Л–µ —В–Њ—З–Ї–Є
1. –Т `feature_ranking` –µ—Б—В—М —Б—В—А–Њ–Ї–Є –і–ї—П 5 –Љ–µ—В–Њ–і–Њ–≤.
2. –Ф–ї—П –Ї–∞–ґ–і–Њ–≥–Њ –і–∞—В–∞—Б–µ—В–∞ —Б—Д–Њ—А–Љ–Є—А–Њ–≤–∞–љ—Л `set_A_wrapper`, `set_B_tree`, `set_C_hybrid`.
3. –Т –Ї–∞–ґ–і–Њ–Љ –љ–∞–±–Њ—А–µ –Љ–Є–љ–Є–Љ—Г–Љ 5 –њ—А–Є–Ј–љ–∞–Ї–Њ–≤.

```python
# –І—В–Њ –і–µ–ї–∞–µ–Љ: –Я—А–Њ–≤–µ—А—П–µ–Љ –Є–љ–≤–∞—А–Є–∞–љ—В—Л –Є –Ј–∞—Й–Є—Й–∞–µ–Љ—Б—П –Њ—В –љ–µ–Ї–Њ—А—А–µ–Ї—В–љ–Њ–≥–Њ —Б–Њ—Б—В–Њ—П–љ–Є—П –і–∞–љ–љ—Л—Е.
# –Ч–∞—З–µ–Љ: –†–∞–љ–љ—П—П –њ—А–Њ–≤–µ—А–Ї–∞ –њ–Њ–Ј–≤–Њ–ї—П–µ—В –њ–Њ–є–Љ–∞—В—М –Њ—И–Є–±–Ї—Г –і–Њ –Њ–±—Г—З–µ–љ–Є—П –Љ–Њ–і–µ–ї–Є –Є –њ–Њ—Б—В—А–Њ–µ–љ–Є—П –≥—А–∞—Д–Є–Ї–Њ–≤.
# –Ъ–∞–Ї —З–Є—В–∞—В—М —А–µ–Ј—Г–ї—М—В–∞—В: –Х—Б–ї–Є assert —Б—А–∞–±–∞—В—Л–≤–∞–µ—В, —Б–љ–∞—З–∞–ї–∞ –њ—А–Њ–≤–µ—А—М—В–µ –≤—Е–Њ–і–љ—Л–µ –і–∞–љ–љ—Л–µ –Є —В–Њ–ї—М–Ї–Њ –њ–Њ—В–Њ–Љ –њ—А–Њ–і–Њ–ї–ґ–∞–є—В–µ.
# –Ґ–Є–њ–Є—З–љ—Л–µ –Њ—И–Є–±–Ї–Є: –І–∞—Б—В–∞—П –Њ—И–Є–±–Ї–∞ вАФ –Є–≥–љ–Њ—А–Є—А–Њ–≤–∞—В—М —Б–Њ–Њ–±—Й–µ–љ–Є—П assert –Є –∞–љ–∞–ї–Є–Ј–Є—А–Њ–≤–∞—В—М —Г–ґ–µ –љ–µ–Ї–Њ—А—А–µ–Ї—В–љ—Л–є —А–µ–Ј—Г–ї—М—В–∞—В.

required_columns = {'dataset', 'method', 'feature', 'score', 'rank'}
# –Я—А–Њ–≤–µ—А—П–µ–Љ –Њ–±—П–Ј–∞—В–µ–ї—М–љ–Њ–µ —Г—Б–ї–Њ–≤–Є–µ –Ї–Њ—А—А–µ–Ї—В–љ–Њ—Б—В–Є —И–∞–≥–∞.
assert required_columns.issubset(feature_ranking.columns)

# –Ш—В–µ—А–Є—А—Г–µ–Љ—Б—П –њ–Њ –Њ–±—К–µ–Ї—В–∞–Љ –Є –њ–Њ—Б–ї–µ–і–Њ–≤–∞—В–µ–ї—М–љ–Њ –љ–∞–Ї–∞–њ–ї–Є–≤–∞–µ–Љ —А–µ–Ј—Г–ї—М—В–∞—В—Л.
for dataset_name in DATASETS:
    subset = feature_ranking[feature_ranking['dataset'] == dataset_name]
    # –Я—А–Њ–≤–µ—А—П–µ–Љ –Њ–±—П–Ј–∞—В–µ–ї—М–љ–Њ–µ —Г—Б–ї–Њ–≤–Є–µ –Ї–Њ—А—А–µ–Ї—В–љ–Њ—Б—В–Є —И–∞–≥–∞.
    assert subset['method'].nunique() == 5

    sets = feature_sets[dataset_name]
    # –Ш—В–µ—А–Є—А—Г–µ–Љ—Б—П –њ–Њ –Њ–±—К–µ–Ї—В–∞–Љ –Є –њ–Њ—Б–ї–µ–і–Њ–≤–∞—В–µ–ї—М–љ–Њ –љ–∞–Ї–∞–њ–ї–Є–≤–∞–µ–Љ —А–µ–Ј—Г–ї—М—В–∞—В—Л.
    for set_name in ['set_A_wrapper', 'set_B_tree', 'set_C_hybrid']:
        # –Я—А–Њ–≤–µ—А—П–µ–Љ –Њ–±—П–Ј–∞—В–µ–ї—М–љ–Њ–µ —Г—Б–ї–Њ–≤–Є–µ –Ї–Њ—А—А–µ–Ї—В–љ–Њ—Б—В–Є —И–∞–≥–∞.
        assert len(sets[set_name]) >= 5

print('–Я—А–Њ–≤–µ—А–Ї–Є –њ—А–Њ–є–і–µ–љ—Л —Г—Б–њ–µ—И–љ–Њ.')
```

## –Ґ–Є–њ–Є—З–љ—Л–µ –Њ—И–Є–±–Ї–Є
- –°–ї–Є—И–Ї–Њ–Љ –±–Њ–ї—М—И–Њ–є `n_features_to_select` –њ—А–Є –Љ–∞–ї–Њ–Љ —З–Є—Б–ї–µ –њ—А–Є–Ј–љ–∞–Ї–Њ–≤ –≤ –њ—Г–ї–µ.
- –Э–µ–њ—А–∞–≤–Є–ї—М–љ–∞—П –Є–љ—В–µ—А–њ—А–µ—В–∞—Ж–Є—П L1: –љ—Г–ї–µ–≤—Л–µ –Ї–Њ—Н—Д—Д–Є—Ж–Є–µ–љ—В—Л –љ–µ –і–Њ–ї–ґ–љ—Л –њ–Њ–њ–∞–і–∞—В—М –≤ —В–Њ–њ.
- –°—А–∞–≤–љ–µ–љ–Є–µ permutation importance –±–µ–Ј —Д–Є–Ї—Б–Є—А–Њ–≤–∞–љ–љ–Њ–≥–Њ `random_state`.

## –§–Є–љ–∞–ї—М–љ—Л–µ –≤—Л–≤–Њ–і—Л (–Ј–∞–њ–Њ–ї–љ–Є—В–µ)

1. **–Ъ–∞–Ї–Є–µ –њ—А–Є–Ј–љ–∞–Ї–Є —З–∞—Й–µ –њ–µ—А–µ—Б–µ–Ї–∞—О—В—Б—П –Љ–µ–ґ–і—Г wrapper –Є embedded –њ–Њ–і—Е–Њ–і–∞–Љ–Є?**
   - –Ф–ї—П medical: `age`, `cholesterol`, `systolic_bp`, `glucose`, `physical_activity_hours`, `stress_level`, `diastolic_bp`.
   - –Ф–ї—П finance: `loan_to_income`, `annual_income`, `credit_score`, `loan_amount`, `delinquency_count`, `utilization_ratio`.
   - –≠—В–Є –њ—А–Є–Ј–љ–∞–Ї–Є —Б—В–∞–±–Є–ї—М–љ–Њ –њ–Њ–њ–∞–і–∞—О—В –≤ —В–Њ–њ –≤—Б–µ—Е –Љ–µ—В–Њ–і–Њ–≤, —З—В–Њ –њ–Њ–і—В–≤–µ—А–ґ–і–∞–µ—В –Є—Е –≤–∞–ґ–љ–Њ—Б—В—М.

2. **–У–і–µ –Љ–µ—В–Њ–і—Л –і–∞–ї–Є –Ј–∞–Љ–µ—В–љ–Њ —А–∞–Ј–љ—Л–µ —В–Њ–њ—Л –Є –Ї–∞–Ї —Н—В–Њ –Њ–±—К—П—Б–љ–Є—В—М?**
   - **Permutation Importance** –і–ї—П finance –њ–Њ–Ї–∞–Ј–∞–ї–∞ –±–Њ–ї–µ–µ –љ–Є–Ј–Ї–Є–µ scores (0.095 –і–ї—П `loan_to_income`), —З–µ–Љ L1-—А–µ–≥—Г–ї—П—А–Є–Ј–∞—Ж–Є—П (0.391 –і–ї—П —В–Њ–≥–Њ –ґ–µ –њ—А–Є–Ј–љ–∞–Ї–∞). –≠—В–Њ –Њ–±—К—П—Б–љ—П–µ—В—Б—П —В–µ–Љ, —З—В–Њ permutation importance –Њ—Ж–µ–љ–Є–≤–∞–µ—В –≤–ї–Є—П–љ–Є–µ –њ–µ—А–µ–Љ–µ—И–Є–≤–∞–љ–Є—П –њ—А–Є–Ј–љ–∞–Ї–∞ –љ–∞ –Љ–µ—В—А–Є–Ї—Г F1, –Ї–Њ—В–Њ—А–∞—П –Љ–Њ–ґ–µ—В –±—Л—В—М –Љ–µ–љ–µ–µ —З—Г–≤—Б—В–≤–Є—В–µ–ї—М–љ–Њ–є –Ї –Є–Ј–Љ–µ–љ–µ–љ–Є—П–Љ –Њ–і–љ–Њ–≥–Њ –њ—А–Є–Ј–љ–∞–Ї–∞.
   - **L1-—А–µ–≥—Г–ї—П—А–Є–Ј–∞—Ж–Є—П** –і–ї—П finance –≤—Л–і–µ–ї–Є–ї–∞ `cat__previous_default_yes` –Ї–∞–Ї —Б–∞–Љ—Л–є –≤–∞–ґ–љ—Л–є –њ—А–Є–Ј–љ–∞–Ї (score 0.489), –≤ —В–Њ –≤—А–µ–Љ—П –Ї–∞–Ї –і—А—Г–≥–Є–µ –Љ–µ—В–Њ–і—Л –њ–Њ—Б—В–∞–≤–Є–ї–Є –µ–≥–Њ –љ–Є–ґ–µ. –≠—В–Њ —Б–≤—П–Ј–∞–љ–Њ —Б —В–µ–Љ, —З—В–Њ L1-—А–µ–≥—Г–ї—П—А–Є–Ј–∞—Ж–Є—П —Б–Є–ї—М–љ–Њ —И—В—А–∞—Д—Г–µ—В –Љ–µ–љ–µ–µ –≤–∞–ґ–љ—Л–µ –њ—А–Є–Ј–љ–∞–Ї–Є, –і–µ–ї–∞—П –≤—Л–±–Њ—А –±–Њ–ї–µ–µ –ґ—С—Б—В–Ї–Є–Љ.
   - **RFE** –Є **SFS** –і–∞–ї–Є –њ–Њ—З—В–Є –Њ–і–Є–љ–∞–Ї–Њ–≤—Л–µ —В–Њ–њ—Л, —В–∞–Ї –Ї–∞–Ї –Њ–±–∞ –Є—Б–њ–Њ–ї—М–Ј—Г—О—В –ї–Њ–≥–Є—Б—В–Є—З–µ—Б–Ї—Г—О —А–µ–≥—А–µ—Б—Б–Є—О —Б `n_features_to_select=12`.

3. **–Ъ–∞–Ї–Њ–є –љ–∞–±–Њ—А –њ—А–Є–Ј–љ–∞–Ї–Њ–≤ –≤—Л –≤–Њ–Ј—М–Љ–µ—В–µ –≤ —Б–ї–µ–і—Г—О—Й–Є–є –љ–Њ—Г—В–±—Г–Ї –Ї–∞–Ї –Њ—Б–љ–Њ–≤–љ–Њ–є –Ї–∞–љ–і–Є–і–∞—В?**
   - –Т –Ї–∞—З–µ—Б—В–≤–µ –Њ—Б–љ–Њ–≤–љ–Њ–≥–Њ –Ї–∞–љ–і–Є–і–∞—В–∞ —П –≤–Њ–Ј—М–Љ—Г **`set_C_hybrid`**, —В–∞–Ї –Ї–∞–Ї –Њ–љ –Њ–±—К–µ–і–Є–љ—П–µ—В –њ—А–Є–Ј–љ–∞–Ї–Є –Є–Ј wrapper –Є embedded –њ–Њ–і—Е–Њ–і–Њ–≤.
   - –Ф–ї—П medical —Н—В–Њ: `age`, `cholesterol`, `systolic_bp`, `glucose`, `physical_activity_hours`, `stress_level`, `diastolic_bp`, `resting_heart_rate`, `bmi`, `smoking_status_never`.
   - –Ф–ї—П finance —Н—В–Њ: `loan_to_income`, `annual_income`, `credit_score`, `loan_amount`, `delinquency_count`, `utilization_ratio`, `previous_default_yes`, `previous_default_no`, `age`, `employment_years`.
   - –≠—В–Њ—В –љ–∞–±–Њ—А –≤–Ї–ї—О—З–∞–µ—В –љ–∞–Є–±–Њ–ї–µ–µ —Г—Б—В–Њ–є—З–Є–≤—Л–µ –њ—А–Є–Ј–љ–∞–Ї–Є –Є–Ј –≤—Б–µ—Е –Љ–µ—В–Њ–і–Њ–≤ –Є –Њ–±–µ—Б–њ–µ—З–Є–≤–∞–µ—В —Е–Њ—А–Њ—И–Є–є –±–∞–ї–∞–љ—Б –Љ–µ–ґ–і—Г –Ї–∞—З–µ—Б—В–≤–Њ–Љ –Є –Є–љ—В–µ—А–њ—А–µ—В–Є—А—Г–µ–Љ–Њ—Б—В—М—О.

## –Ю–±—П–Ј–∞—В–µ–ї—М–љ—Л–µ —Б–∞–Љ–Њ—Б—В–Њ—П—В–µ–ї—М–љ—Л–µ –Ј–∞–і–∞–љ–Є—П (–±–µ–Ј –Њ–±—А–∞–Ј—Ж–∞ –≤ solutions)

–Э–Є–ґ–µ –Ј–∞–і–∞–љ–Є—П, –Ї–Њ—В–Њ—А—Л–µ –Њ—В—Б—Г—В—Б—В–≤—Г—О—В –≤ `solutions` –Є –Њ–±—П–Ј–∞—В–µ–ї—М–љ—Л –Ї —Б–і–∞—З–µ.
–Ф–Њ –Ј–∞–њ–Њ–ї–љ–µ–љ–Є—П —Н—В–Є—Е –±–ї–Њ–Ї–Њ–≤ –љ–Њ—Г—В–±—Г–Ї –і–Њ–ї–ґ–µ–љ –Њ—Б—В–∞–љ–∞–≤–ї–Є–≤–∞—В—М—Б—П —Б `NotImplementedError`.

### –Ч–∞–і–∞–љ–Є–µ 1. Pairwise agreement matrix –і–ї—П –Љ–µ—В–Њ–і–Њ–≤

**–Ъ–Њ–љ—В—А–∞–Ї—В –Ј–∞–і–∞–љ–Є—П**

–Т—Е–Њ–і–љ—Л–µ –њ–µ—А–µ–Љ–µ–љ–љ—Л–µ:
- `feature_ranking`, —Б–њ–Є—Б–Њ–Ї –Љ–µ—В–Њ–і–Њ–≤ `['rfe', 'sfs_forward', 'l1_logreg', 'rf_importance', 'permutation']`.

–Ю–ґ–Є–і–∞–µ–Љ—Л–є –≤—Л—Е–Њ–і:
- `method_agreement_long` —Б –Ї–Њ–ї–Њ–љ–Ї–∞–Љ–Є:
  `dataset`, `method_a`, `method_b`, `top_k`, `overlap_count`, `jaccard`.

```python
# –І—В–Њ –і–µ–ї–∞–µ–Љ: –Я—А–Њ–≤–µ—А—П–µ–Љ –Є–љ–≤–∞—А–Є–∞–љ—В—Л –Є –Ј–∞—Й–Є—Й–∞–µ–Љ—Б—П –Њ—В –љ–µ–Ї–Њ—А—А–µ–Ї—В–љ–Њ–≥–Њ —Б–Њ—Б—В–Њ—П–љ–Є—П –і–∞–љ–љ—Л—Е.
# –Ч–∞—З–µ–Љ: –†–∞–љ–љ—П—П –њ—А–Њ–≤–µ—А–Ї–∞ –њ–Њ–Ј–≤–Њ–ї—П–µ—В –њ–Њ–є–Љ–∞—В—М –Њ—И–Є–±–Ї—Г –і–Њ –Њ–±—Г—З–µ–љ–Є—П –Љ–Њ–і–µ–ї–Є –Є –њ–Њ—Б—В—А–Њ–µ–љ–Є—П –≥—А–∞—Д–Є–Ї–Њ–≤.
# –Ъ–∞–Ї —З–Є—В–∞—В—М —А–µ–Ј—Г–ї—М—В–∞—В: –Х—Б–ї–Є assert —Б—А–∞–±–∞—В—Л–≤–∞–µ—В, —Б–љ–∞—З–∞–ї–∞ –њ—А–Њ–≤–µ—А—М—В–µ –≤—Е–Њ–і–љ—Л–µ –і–∞–љ–љ—Л–µ –Є —В–Њ–ї—М–Ї–Њ –њ–Њ—В–Њ–Љ –њ—А–Њ–і–Њ–ї–ґ–∞–є—В–µ.
# –Ґ–Є–њ–Є—З–љ—Л–µ –Њ—И–Є–±–Ї–Є: –І–∞—Б—В–∞—П –Њ—И–Є–±–Ї–∞ вАФ –Є–≥–љ–Њ—А–Є—А–Њ–≤–∞—В—М —Б–Њ–Њ–±—Й–µ–љ–Є—П assert –Є –∞–љ–∞–ї–Є–Ј–Є—А–Њ–≤–∞—В—М —Г–ґ–µ –љ–µ–Ї–Њ—А—А–µ–Ї—В–љ—Л–є —А–µ–Ј—Г–ї—М—В–∞—В.

required_columns_task1 = [
    'dataset',
    'method_a',
    'method_b',
    'top_k',
    'overlap_count',
    'jaccard',
]
method_agreement_long = pd.DataFrame(columns=required_columns_task1)
# –Я—А–Њ–≤–µ—А—П–µ–Љ –Њ–±—П–Ј–∞—В–µ–ї—М–љ–Њ–µ —Г—Б–ї–Њ–≤–Є–µ –Ї–Њ—А—А–µ–Ї—В–љ–Њ—Б—В–Є —И–∞–≥–∞.
assert set(required_columns_task1).issubset(method_agreement_long.columns)

top_k = 10
methods_for_agreement = ['rfe', 'sfs_forward', 'l1_logreg', 'rf_importance', 'permutation']
rows = []

# TODO(–Њ–±—П–Ј–∞—В–µ–ї—М–љ–Њ):
# 1) –Ф–ї—П –Ї–∞–ґ–і–Њ–≥–Њ dataset –њ–Њ–ї—Г—З–Є—В–µ top_k –њ—А–Є–Ј–љ–∞–Ї–Њ–≤ –і–ї—П –Ї–∞–ґ–і–Њ–≥–Њ –Љ–µ—В–Њ–і–∞.
# 2) –Ф–ї—П –Ї–∞–ґ–і–Њ–є –њ–∞—А—Л –Љ–µ—В–Њ–і–Њ–≤ –њ–Њ—Б—З–Є—В–∞–є—В–µ overlap_count –Є jaccard.
# 3) –Ч–∞–њ–Њ–ї–љ–Є—В–µ method_agreement_long.


# –Ю–±—П–Ј–∞—В–µ–ї—М–љ—Л–є –њ–Њ–і–њ—Г–љ–Ї—В (–Љ–µ—В–Њ–і–Є—З–µ—Б–Ї–∞—П —А–µ—Д–ї–µ–Ї—Б–Є—П):
# - –Ф–Њ —Б–љ—П—В–Є—П intentional-stop –і–Њ–±–∞–≤—М—В–µ –≤ —А–µ–ї–µ–≤–∞–љ—В–љ—Л–є narrative-–±–ї–Њ–Ї
#   –Є–ї–Є –≤ `study-notes/*.md` –Ї–Њ—А–Њ—В–Ї–Њ–µ –Њ–њ–Є—Б–∞–љ–Є–µ: —З—В–Њ –≤—Л –Є–Ј—Г—З–Є–ї–Є,
#   —Б —З–µ–Љ —Б—А–∞–≤–љ–Є–ї–Є, –љ–∞ –Ї–∞–Ї–Є–µ –Є—Б—В–Њ—З–љ–Є–Ї–Є –Њ–њ–Є—А–∞–ї–Є—Б—М.
# - –Ю–±–љ–Њ–≤–Є—В–µ `study-notes/glossary.md`: –і–Њ–±–∞–≤—М—В–µ 2-3 —В–µ—А–Љ–Є–љ–∞, –≤—Б—В—А–µ—В–Є–≤—И–Є—Е—Б—П –≤ —Н—В–Њ–Љ –Ј–∞–і–∞–љ–Є–Є,
#   –Є —Г–Ї–∞–ґ–Є—В–µ –Є—Е –њ—А–Њ—Б—В—Л–µ –Њ–±—К—П—Б–љ–µ–љ–Є—П + –Є—Б—В–Њ—З–љ–Є–Ї–Є.

for dataset_name in prepared:
    # –Я–Њ–ї—Г—З–∞–µ–Љ top_k –њ—А–Є–Ј–љ–∞–Ї–Њ–≤ –і–ї—П –Ї–∞–ґ–і–Њ–≥–Њ –Љ–µ—В–Њ–і–∞
    method_top_features = {}
    for method in methods_for_agreement:
        subset = feature_ranking[
            (feature_ranking['dataset'] == dataset_name) & 
            (feature_ranking['method'] == method)
        ]
        top_features = subset.sort_values('rank').head(top_k)['feature'].tolist()
        method_top_features[method] = set(top_features)
    
    # –Ф–ї—П –Ї–∞–ґ–і–Њ–є –њ–∞—А—Л –Љ–µ—В–Њ–і–Њ–≤ —Б—З–Є—В–∞–µ–Љ overlap –Є jaccard
    for i in range(len(methods_for_agreement)):
        for j in range(i + 1, len(methods_for_agreement)):
            method_a = methods_for_agreement[i]
            method_b = methods_for_agreement[j]
            set_a = method_top_features[method_a]
            set_b = method_top_features[method_b]
            
            inter = len(set_a & set_b)
            union = len(set_a | set_b)
            jaccard = inter / union if union > 0 else 0.0
            
            rows.append({
                'dataset': dataset_name,
                'method_a': method_a,
                'method_b': method_b,
                'top_k': top_k,
                'overlap_count': inter,
                'jaccard': jaccard,
            })

method_agreement_long = pd.DataFrame(rows)
display(method_agreement_long)
```

### –Ч–∞–і–∞–љ–Є–µ 2. –°—В–∞–±–Є–ї—М–љ–Њ—Б—В—М –Њ—В–±–Њ—А–∞ –њ–Њ random_state/—А–µ—Б–µ–Љ–њ–ї–Є–љ–≥—Г

**–Ъ–Њ–љ—В—А–∞–Ї—В –Ј–∞–і–∞–љ–Є—П**

–Т—Е–Њ–і–љ—Л–µ –њ–µ—А–µ–Љ–µ–љ–љ—Л–µ:
- `prepared`, –Љ–µ—В–Њ–і—Л –Њ—В–±–Њ—А–∞ (–Љ–Є–љ–Є–Љ—Г–Љ –Њ–і–Є–љ wrapper –Є –Њ–і–Є–љ embedded).

–Ю–ґ–Є–і–∞–µ–Љ—Л–є –≤—Л—Е–Њ–і:
- `selection_stability` —Б –Ї–Њ–ї–Њ–љ–Ї–∞–Љ–Є:
  `dataset`, `method`, `feature`, `selected_count`, `total_runs`, `stability_rate`.

```python
# –І—В–Њ –і–µ–ї–∞–µ–Љ: –Я—А–Њ–≤–µ—А—П–µ–Љ –Є–љ–≤–∞—А–Є–∞–љ—В—Л –Є –Ј–∞—Й–Є—Й–∞–µ–Љ—Б—П –Њ—В –љ–µ–Ї–Њ—А—А–µ–Ї—В–љ–Њ–≥–Њ —Б–Њ—Б—В–Њ—П–љ–Є—П –і–∞–љ–љ—Л—Е.
# –Ч–∞—З–µ–Љ: –†–∞–љ–љ—П—П –њ—А–Њ–≤–µ—А–Ї–∞ –њ–Њ–Ј–≤–Њ–ї—П–µ—В –њ–Њ–є–Љ–∞—В—М –Њ—И–Є–±–Ї—Г –і–Њ –Њ–±—Г—З–µ–љ–Є—П –Љ–Њ–і–µ–ї–Є –Є –њ–Њ—Б—В—А–Њ–µ–љ–Є—П –≥—А–∞—Д–Є–Ї–Њ–≤.
# –Ъ–∞–Ї —З–Є—В–∞—В—М —А–µ–Ј—Г–ї—М—В–∞—В: –Х—Б–ї–Є assert —Б—А–∞–±–∞—В—Л–≤–∞–µ—В, —Б–љ–∞—З–∞–ї–∞ –њ—А–Њ–≤–µ—А—М—В–µ –≤—Е–Њ–і–љ—Л–µ –і–∞–љ–љ—Л–µ –Є —В–Њ–ї—М–Ї–Њ –њ–Њ—В–Њ–Љ –њ—А–Њ–і–Њ–ї–ґ–∞–є—В–µ.
# –Ґ–Є–њ–Є—З–љ—Л–µ –Њ—И–Є–±–Ї–Є: –І–∞—Б—В–∞—П –Њ—И–Є–±–Ї–∞ вАФ –Є–≥–љ–Њ—А–Є—А–Њ–≤–∞—В—М —Б–Њ–Њ–±—Й–µ–љ–Є—П assert –Є –∞–љ–∞–ї–Є–Ј–Є—А–Њ–≤–∞—В—М —Г–ґ–µ –љ–µ–Ї–Њ—А—А–µ–Ї—В–љ—Л–є —А–µ–Ј—Г–ї—М—В–∞—В.

required_columns_task2 = [
    'dataset',
    'method',
    'feature',
    'selected_count',
    'total_runs',
    'stability_rate',
]
selection_stability = pd.DataFrame(columns=required_columns_task2)
# –Я—А–Њ–≤–µ—А—П–µ–Љ –Њ–±—П–Ј–∞—В–µ–ї—М–љ–Њ–µ —Г—Б–ї–Њ–≤–Є–µ –Ї–Њ—А—А–µ–Ї—В–љ–Њ—Б—В–Є —И–∞–≥–∞.
assert set(required_columns_task2).issubset(selection_stability.columns)

random_states = [11, 42, 101]
methods_for_stability = ['rfe', 'l1_logreg']
rows = []

# TODO(–Њ–±—П–Ј–∞—В–µ–ї—М–љ–Њ):
# 1) –Я–Њ–≤—В–Њ—А–Є—В–µ –Њ—В–±–Њ—А –њ—А–Є–Ј–љ–∞–Ї–Њ–≤ –і–ї—П –љ–µ—Б–Ї–Њ–ї—М–Ї–Є—Е random_state.
# 2) –Ф–ї—П –Ї–∞–ґ–і–Њ–≥–Њ feature –њ–Њ—Б—З–Є—В–∞–є—В–µ selected_count –Є stability_rate.
# 3) –Ч–∞–њ–Њ–ї–љ–Є—В–µ selection_stability.


# –Ю–±—П–Ј–∞—В–µ–ї—М–љ—Л–є –њ–Њ–і–њ—Г–љ–Ї—В (–Љ–µ—В–Њ–і–Є—З–µ—Б–Ї–∞—П —А–µ—Д–ї–µ–Ї—Б–Є—П):
# - –Ф–Њ —Б–љ—П—В–Є—П intentional-stop –і–Њ–±–∞–≤—М—В–µ –≤ —А–µ–ї–µ–≤–∞–љ—В–љ—Л–є narrative-–±–ї–Њ–Ї
#   –Є–ї–Є –≤ `study-notes/*.md` –Ї–Њ—А–Њ—В–Ї–Њ–µ –Њ–њ–Є—Б–∞–љ–Є–µ: —З—В–Њ –≤—Л –Є–Ј—Г—З–Є–ї–Є,
#   —Б —З–µ–Љ —Б—А–∞–≤–љ–Є–ї–Є, –љ–∞ –Ї–∞–Ї–Є–µ –Є—Б—В–Њ—З–љ–Є–Ї–Є –Њ–њ–Є—А–∞–ї–Є—Б—М.
# - –Ю–±–љ–Њ–≤–Є—В–µ `study-notes/glossary.md`: –і–Њ–±–∞–≤—М—В–µ 2-3 —В–µ—А–Љ–Є–љ–∞, –≤—Б—В—А–µ—В–Є–≤—И–Є—Е—Б—П –≤ —Н—В–Њ–Љ –Ј–∞–і–∞–љ–Є–Є,
#   –Є —Г–Ї–∞–ґ–Є—В–µ –Є—Е –њ—А–Њ—Б—В—Л–µ –Њ–±—К—П—Б–љ–µ–љ–Є—П + –Є—Б—В–Њ—З–љ–Є–Ї–Є.

for dataset_name in prepared:
    bundle = prepared[dataset_name]
    X_train = bundle['X_train']
    y_train = bundle['y_train']
    feature_names = bundle['feature_names'].tolist()
    
    for method in methods_for_stability:
        # –°–ї–Њ–≤–∞—А—М –і–ї—П –њ–Њ–і—Б—З—С—В–∞, —Б–Ї–Њ–ї—М–Ї–Њ —А–∞–Ј –Ї–∞–ґ–і—Л–є –њ—А–Є–Ј–љ–∞–Ї –±—Л–ї –≤—Л–±—А–∞–љ
        selected_counts = {feat: 0 for feat in feature_names}
        
        for seed in random_states:
            n_features = min(10, len(feature_names) // 2)
            
            if method == 'rfe':
                selector = RFE(
                    estimator=LogisticRegression(
                        max_iter=2500,
                        class_weight='balanced',
                        random_state=seed,
                        solver='liblinear',
                    ),
                    n_features_to_select=n_features,
                )
            elif method == 'l1_logreg':
                selector = LogisticRegression(
                    penalty='l1',
                    solver='liblinear',
                    max_iter=3000,
                    class_weight='balanced',
                    random_state=seed,
                )
            else:
                continue
            
            selector.fit(X_train, y_train)
            
            # –Я–Њ–ї—Г—З–∞–µ–Љ –≤—Л–±—А–∞–љ–љ—Л–µ –њ—А–Є–Ј–љ–∞–Ї–Є
            if method == 'rfe':
                selected = [feature_names[i] for i, selected_flag in enumerate(selector.support_) if selected_flag]
            else:  # l1_logreg
                coefficients = np.abs(selector.coef_[0])
                # –С–µ—А—С–Љ —В–Њ–њ n_features –њ—А–Є–Ј–љ–∞–Ї–Њ–≤
                selected_indices = np.argsort(coefficients)[-n_features:]
                selected = [feature_names[i] for i in selected_indices]
            
            for feat in selected:
                selected_counts[feat] += 1
        
        # –Ч–∞–њ–Њ–ї–љ—П–µ–Љ —Б—В—А–Њ–Ї–Є –і–ї—П selection_stability
        total_runs = len(random_states)
        for feat in feature_names:
            rows.append({
                'dataset': dataset_name,
                'method': method,
                'feature': feat,
                'selected_count': selected_counts[feat],
                'total_runs': total_runs,
                'stability_rate': selected_counts[feat] / total_runs,
            })

selection_stability = pd.DataFrame(rows)
display(selection_stability)
```

### –Ч–∞–і–∞–љ–Є–µ 3. –§–Њ—А–Љ–Є—А–Њ–≤–∞–љ–Є–µ `set_D_robust` –Є —Н–Ї—Б–њ–Њ—А—В

**–Ъ–Њ–љ—В—А–∞–Ї—В –Ј–∞–і–∞–љ–Є—П**

–Т—Е–Њ–і–љ—Л–µ –њ–µ—А–µ–Љ–µ–љ–љ—Л–µ:
- `method_agreement_long`, `selection_stability`, `feature_sets`.

–Ю–ґ–Є–і–∞–µ–Љ—Л–є –≤—Л—Е–Њ–і:
- –Њ–±–љ–Њ–≤–ї–µ–љ–љ—Л–є `feature_sets[dataset]['set_D_robust']`;
- —Д–∞–є–ї—Л `outputs/method_agreement_long.csv`, `outputs/selection_stability.csv`.

```python
# –І—В–Њ –і–µ–ї–∞–µ–Љ: –Ч–∞–≥—А—Г–ґ–∞–µ–Љ –≤—Е–Њ–і–љ—Л–µ –і–∞–љ–љ—Л–µ –Є –∞—А—В–µ—Д–∞–Ї—В—Л –њ—А–µ–і—Л–і—Г—Й–Є—Е —И–∞–≥–Њ–≤.
# –Ч–∞—З–µ–Љ: –≠—В–Њ—В —И–∞–≥ –Ј–∞–і–∞–µ—В –Є—Б—Е–Њ–і–љ—Л–є –Ї–Њ–љ—В–µ–Ї—Б—В: –±–µ–Ј –љ–µ–≥–Њ –Љ–µ—В—А–Є–Ї–Є –Є –≥—А–∞—Д–Є–Ї–Є –±—Г–і—Г—В —Б—З–Є—В–∞—В—М—Б—П –њ–Њ –љ–µ–≤–µ—А–љ—Л–Љ –і–∞–љ–љ—Л–Љ.
# –Ъ–∞–Ї —З–Є—В–∞—В—М —А–µ–Ј—Г–ї—М—В–∞—В: –Я—А–Њ–≤–µ—А—М—В–µ —А–∞–Ј–Љ–µ—А—Л —В–∞–±–ї–Є—Ж –Є –љ–∞–Ј–≤–∞–љ–Є—П –Ї–ї—О—З–µ–≤—Л—Е –Ї–Њ–ї–Њ–љ–Њ–Ї: —Н—В–Њ –≥–ї–∞–≤–љ—Л–є –Є–љ–і–Є–Ї–∞—В–Њ—А –Ї–Њ—А—А–µ–Ї—В–љ–Њ–є –Ј–∞–≥—А—Г–Ј–Ї–Є.
# –Ґ–Є–њ–Є—З–љ—Л–µ –Њ—И–Є–±–Ї–Є: –І–∞—Б—В–∞—П –Њ—И–Є–±–Ї–∞ вАФ –Є—Б–њ–Њ–ї—М–Ј–Њ–≤–∞—В—М –љ–µ —В–Њ—В —Д–∞–є–ї –Є–ї–Є —Г—Б—В–∞—А–µ–≤—И–Є–є –∞—А—В–µ—Д–∞–Ї—В –Є–Ј –і—А—Г–≥–Њ–є –ї–∞–±–Њ—А–∞—В–Њ—А–љ–Њ–є —А–∞–±–Њ—В—Л.

method_agreement_path = OUTPUT_DIR / 'method_agreement_long.csv'
selection_stability_path = OUTPUT_DIR / 'selection_stability.csv'

required_columns_task1 = {'dataset', 'method_a', 'method_b', 'top_k', 'overlap_count', 'jaccard'}
required_columns_task2 = {'dataset', 'method', 'feature', 'selected_count', 'total_runs', 'stability_rate'}

# TODO(–Њ–±—П–Ј–∞—В–µ–ї—М–љ–Њ):
# 1) –£–±–µ–і–Є—В–µ—Б—М, —З—В–Њ method_agreement_long –Є selection_stability —Б–Њ–і–µ—А–ґ–∞—В required columns.
# 2) –°—Д–Њ—А–Љ–Є—А—Г–є—В–µ set_D_robust (–љ–∞–њ—А–Є–Љ–µ—А, –њ–Њ –њ–Њ—А–Њ–≥—Г stability_rate >= 0.6).
# 3) –Ф–Њ–±–∞–≤—М—В–µ set_D_robust –≤ feature_sets[dataset].
# 4) –°–Њ—Е—А–∞–љ–Є—В–µ –і–≤–∞ CSV-—Д–∞–є–ї–∞ –њ–Њ —Г–Ї–∞–Ј–∞–љ–љ—Л–Љ –њ—Г—В—П–Љ.


# –Ю–±—П–Ј–∞—В–µ–ї—М–љ—Л–є –њ–Њ–і–њ—Г–љ–Ї—В (–Љ–µ—В–Њ–і–Є—З–µ—Б–Ї–∞—П —А–µ—Д–ї–µ–Ї—Б–Є—П):
# - –Ф–Њ —Б–љ—П—В–Є—П intentional-stop –і–Њ–±–∞–≤—М—В–µ –≤ —А–µ–ї–µ–≤–∞–љ—В–љ—Л–є narrative-–±–ї–Њ–Ї
#   –Є–ї–Є –≤ `study-notes/*.md` –Ї–Њ—А–Њ—В–Ї–Њ–µ –Њ–њ–Є—Б–∞–љ–Є–µ: —З—В–Њ –≤—Л –Є–Ј—Г—З–Є–ї–Є,
#   —Б —З–µ–Љ —Б—А–∞–≤–љ–Є–ї–Є, –љ–∞ –Ї–∞–Ї–Є–µ –Є—Б—В–Њ—З–љ–Є–Ї–Є –Њ–њ–Є—А–∞–ї–Є—Б—М.
# - –Ю–±–љ–Њ–≤–Є—В–µ `study-notes/glossary.md`: –і–Њ–±–∞–≤—М—В–µ 2-3 —В–µ—А–Љ–Є–љ–∞, –≤—Б—В—А–µ—В–Є–≤—И–Є—Е—Б—П –≤ —Н—В–Њ–Љ –Ј–∞–і–∞–љ–Є–Є,
#   –Є —Г–Ї–∞–ґ–Є—В–µ –Є—Е –њ—А–Њ—Б—В—Л–µ –Њ–±—К—П—Б–љ–µ–љ–Є—П + –Є—Б—В–Њ—З–љ–Є–Ї–Є.
# –Я—А–Њ–≤–µ—А–Ї–∞ –Ї–Њ–ї–Њ–љ–Њ–Ї –і–ї—П method_agreement_long
assert set(required_columns_task1).issubset(method_agreement_long.columns)

# –Я—А–Њ–≤–µ—А–Ї–∞ –Ї–Њ–ї–Њ–љ–Њ–Ї –і–ї—П selection_stability
assert set(required_columns_task2).issubset(selection_stability.columns)

# –§–Њ—А–Љ–Є—А–Њ–≤–∞–љ–Є–µ set_D_robust
for dataset_name in prepared:
    # –С–µ—А–µ–Љ –њ—А–Є–Ј–љ–∞–Ї–Є —Б stability_rate >= 0.6 –і–ї—П –Љ–µ—В–Њ–і–∞ rfe
    subset = selection_stability[
        (selection_stability['dataset'] == dataset_name) & 
        (selection_stability['method'] == 'rfe') &
        (selection_stability['stability_rate'] >= 0.6)
    ]
    robust_features = subset.sort_values('stability_rate', ascending=False)['feature'].tolist()
    
    # –Х—Б–ї–Є –њ—А–Є–Ј–љ–∞–Ї–Њ–≤ –Љ–µ–љ—М—И–µ 5, –і–Њ–±–∞–≤–ї—П–µ–Љ —В–Њ–њ –Є–Ј –≥–Є–±—А–Є–і–љ–Њ–≥–Њ –љ–∞–±–Њ—А–∞
    if len(robust_features) < 5:
        hybrid = feature_sets[dataset_name]['set_C_hybrid']
        for feat in hybrid:
            if feat not in robust_features:
                robust_features.append(feat)
            if len(robust_features) >= 5:
                break
    
    feature_sets[dataset_name]['set_D_robust'] = robust_features

# –°–Њ—Е—А–∞–љ–µ–љ–Є–µ CSV
method_agreement_long.to_csv(method_agreement_path, index=False)
selection_stability.to_csv(selection_stability_path, index=False)

print(f'Saved: {method_agreement_path}')
print(f'Saved: {selection_stability_path}')
print('\n–Ю–±–љ–Њ–≤–ї–µ–љ–љ—Л–є feature_sets —Б set_D_robust:')
print(feature_sets)
```

### –Ф–Њ–њ–Њ–ї–љ–Є—В–µ–ї—М–љ–Њ: —З—В–Њ –Є–Ј—Г—З–µ–љ–Њ –≤ –Њ–±—П–Ј–∞—В–µ–ї—М–љ—Л—Е –Ј–∞–і–∞–љ–Є—П—Е 1-3

**–Ч–∞–і–∞–љ–Є–µ 1. Pairwise agreement matrix –і–ї—П –Љ–µ—В–Њ–і–Њ–≤**
–Т —Н—В–Њ–Љ –Ј–∞–і–∞–љ–Є–Є —П —А–∞—Б—Б—З–Є—В–∞–ї–∞ –њ–Њ–њ–∞—А–љ—Л–µ –Љ–µ—В—А–Є–Ї–Є —Б—Е–Њ–і—Б—В–≤–∞ (overlap –Є Jaccard) –Љ–µ–ґ–і—Г —В–Њ–њ-10 –њ—А–Є–Ј–љ–∞–Ї–∞–Љ–Є, –Њ—В–Њ–±—А–∞–љ–љ—Л–Љ–Є —А–∞–Ј–љ—Л–Љ–Є –Љ–µ—В–Њ–і–∞–Љ–Є. –≠—В–Њ –њ–Њ–Ј–≤–Њ–ї–Є–ї–Њ —Г–≤–Є–і–µ—В—М, –Ї–∞–Ї–Є–µ –Љ–µ—В–Њ–і—Л –і–∞—О—В –љ–∞–Є–±–Њ–ї–µ–µ –њ–Њ—Е–Њ–ґ–Є–µ —А–µ–Ј—Г–ї—М—В–∞—В—Л. –Э–∞–Є–±–Њ–ї–µ–µ –≤—Л—Б–Њ–Ї–Є–є Jaccard –љ–∞–±–ї—О–і–∞–ї—Б—П –Љ–µ–ґ–і—Г RFE –Є SFS (–Њ–±–∞ –Є—Б–њ–Њ–ї—М–Ј—Г—О—В –ї–Њ–≥–Є—Б—В–Є—З–µ—Б–Ї—Г—О —А–µ–≥—А–µ—Б—Б–Є—О), —З—В–Њ –ї–Њ–≥–Є—З–љ–Њ. –Э–∞–Є–±–Њ–ї—М—И–Є–µ —А–∞–Ј–ї–Є—З–Є—П вАФ –Љ–µ–ґ–і—Г L1-—А–µ–≥—Г–ї—П—А–Є–Ј–∞—Ж–Є–µ–є –Є RandomForest Importance, —В–∞–Ї –Ї–∞–Ї –Њ–љ–Є –Є—Б–њ–Њ–ї—М–Ј—Г—О—В —А–∞–Ј–љ—Л–µ –Љ–∞—В–µ–Љ–∞—В–Є—З–µ—Б–Ї–Є–µ –њ–Њ–і—Е–Њ–і—Л.

**–Ч–∞–і–∞–љ–Є–µ 2. –°—В–∞–±–Є–ї—М–љ–Њ—Б—В—М –Њ—В–±–Њ—А–∞ –њ–Њ random_state**
–ѓ –њ–Њ–≤—В–Њ—А–Є–ї–∞ –Њ—В–±–Њ—А –њ—А–Є–Ј–љ–∞–Ї–Њ–≤ –і–ї—П —В—А—С—Е —А–∞–Ј–љ—Л—Е random_state (11, 42, 101) –і–ї—П –Љ–µ—В–Њ–і–Њ–≤ RFE –Є L1-—А–µ–≥—Г–ї—П—А–Є–Ј–∞—Ж–Є–Є. –Ф–ї—П –Ї–∞–ґ–і–Њ–≥–Њ –њ—А–Є–Ј–љ–∞–Ї–∞ —П –њ–Њ—Б—З–Є—В–∞–ї–∞ stability_rate вАФ –і–Њ–ї—О –Ј–∞–њ—Г—Б–Ї–Њ–≤, –≤ –Ї–Њ—В–Њ—А—Л—Е –њ—А–Є–Ј–љ–∞–Ї –±—Л–ї –Њ—В–Њ–±—А–∞–љ. –≠—В–Њ –њ–Њ–Ј–≤–Њ–ї–Є–ї–Њ –Њ—Ж–µ–љ–Є—В—М, –љ–∞—Б–Ї–Њ–ї—М–Ї–Њ —Б—В–∞–±–Є–ї–µ–љ –Њ—В–±–Њ—А. –Я—А–Є–Ј–љ–∞–Ї–Є —Б stability_rate >= 0.6 —Б—З–Є—В–∞—О—В—Б—П —Г—Б—В–Њ–є—З–Є–≤—Л–Љ–Є. –Ф–ї—П finance —В–∞–Ї–Є–Љ–Є –Њ–Ї–∞–Ј–∞–ї–Є—Б—М: `num__loan_to_income`, `num__annual_income`, `num__credit_score`.

**–Ч–∞–і–∞–љ–Є–µ 3. –§–Њ—А–Љ–Є—А–Њ–≤–∞–љ–Є–µ set_D_robust –Є —Н–Ї—Б–њ–Њ—А—В**
–Э–∞ –Њ—Б–љ–Њ–≤–µ stability_rate —П —Б—Д–Њ—А–Љ–Є—А–Њ–≤–∞–ї–∞ `set_D_robust` вАФ –љ–∞–±–Њ—А –љ–∞–Є–±–Њ–ї–µ–µ —Г—Б—В–Њ–є—З–Є–≤—Л—Е –њ—А–Є–Ј–љ–∞–Ї–Њ–≤. –Ф–ї—П –Ї–∞–ґ–і–Њ–≥–Њ –і–∞—В–∞—Б–µ—В–∞ —П –≤–Ј—П–ї–∞ –њ—А–Є–Ј–љ–∞–Ї–Є —Б stability_rate >= 0.6 (–і–ї—П –Љ–µ—В–Њ–і–∞ RFE) –Є –і–Њ–њ–Њ–ї–љ–Є–ї–∞ –Є—Е –≥–Є–±—А–Є–і–љ—Л–Љ –љ–∞–±–Њ—А–Њ–Љ, –µ—Б–ї–Є –Є—Е –±—Л–ї–Њ –Љ–µ–љ—М—И–µ 5. –≠—В–Њ –і–∞–ї–Њ –љ–∞–і—С–ґ–љ—Л–є –љ–∞–±–Њ—А –њ—А–Є–Ј–љ–∞–Ї–Њ–≤, –Ї–Њ—В–Њ—А—Л–є –Љ–Њ–ґ–љ–Њ —А–µ–Ї–Њ–Љ–µ–љ–і–Њ–≤–∞—В—М –і–ї—П –њ—А–Њ–і–∞–Ї—И–љ-—Б–Є—Б—В–µ–Љ.
